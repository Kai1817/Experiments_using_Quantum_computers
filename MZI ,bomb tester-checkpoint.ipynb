```python
from qiskit import*
```

```python
bomb_tester = QuantumCircuit(2)
```

```python
%matplotlib inline 
bomb_tester.draw(output='mpl')
```

```python
bomb_tester.h(0)
bomb_tester.cx(0,1)
bomb_tester.h(0)
bomb_tester.measure_all()
```

```python
bomb_tester.draw(output='mpl')
```

```python
#Running the bomb tester circuit
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = simulator.run(bomb_tester , shots=1000)
result = job.result()
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(result.get_counts(bomb_tester))
```
